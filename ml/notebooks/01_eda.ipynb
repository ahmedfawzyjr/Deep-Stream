# DeepStream — Match Prediction EDA
Exploratory Data Analysis of StatsBomb La Liga match event statistics.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load preprocessed match features
df = pd.read_parquet('../data/processed/features.parquet')
df.head()
```

## Match Outcomes Distribution (0=Away Win, 1=Draw, 2=Home Win)

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='result', data=df, palette='viridis')
plt.title('Distribution of Match Outcomes')
plt.xlabel('Result Label')
plt.ylabel('Count')
plt.show()
```

## Feature Distributions & Correlations

```python
feature_cols = [
    'home_shots', 'away_shots',
    'home_shots_on_target', 'away_shots_on_target',
    'home_pass_accuracy', 'away_pass_accuracy',
    'home_possession_pct', 'away_possession_pct',
    'home_xg', 'away_xg',
    'home_form_last5', 'away_form_last5',
    'home_goals_scored_avg', 'away_goals_scored_avg',
    'head_to_head_home_wins'
]

# Compute correlation matrix
corr = df[feature_cols].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Feature Correlation Matrix')
plt.show()
```
